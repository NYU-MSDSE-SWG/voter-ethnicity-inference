```python
df = df[['cbg2000','total','white','black','api','aian','2race','hispanic']]
```

```python
df['white'] = df['white']/df['total']
```

```python
df['black'] = df['black']/df['total']
df['api'] = df['api']/df['total']
df['aian'] =df['aian']/df['total']
df['2race'] =df['2race']/df['total']
df['hispanic'] = df['hispanic']/df['total']
```
